```python
#Import the necessary libraries
import pandas as pd
import re
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
import gensim
from nltk.tokenize import word_tokenize
```

```python
#download the stopwords 
import nltk
nltk.download('stopwords')
```

```python
df = pd.read_csv("../Data/courses.csv")
df
```

```python
#Define a function to clean non-alphanumeric characters and tokenize the text
def clean_and_tokenize(text):
    # Remove non-alphanumeric characters and convert to lowercase
    text = re.sub(r'[^a-zA-Z0-9 ]', '', str(text).lower())
    # Tokenize the text
    tokens = text.split()
    return ' '.join(tokens)
```

```python
# Apply the cleaning and tokenization function to all columns in the DataFrame
for column in df.columns:
    df[column] = df[column].apply(clean_and_tokenize)
```

```python
df
```

```python
#Define a list of stopwords and create a CountVectorizer to convert the text data into a Bag of Words
stop_words = set(stopwords.words('english'))
vectorizer = CountVectorizer(stop_words=stop_words)
```

```python
#Fit and transform the text data using the CountVectorizer
text_data = df.apply(lambda row: ' '.join(row), axis=1)
bag_of_words = vectorizer.fit_transform(text_data)
```

```python
#Convert the Bag of Words into a DataFrame
bow_df = pd.DataFrame(bag_of_words.toarray(), columns=vectorizer.get_feature_names_out())
```

```python
bow_df
```

```python
def tokenize_course(course, keep_only_nouns=True):
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(course)
    # Remove English stop words
    word_tokens = [w for w in word_tokens if (not w.lower() in stop_words)]
    return word_tokens
```

```python
tokenized_courses = [tokenize_course(course_text) for course_text in text_data]
tokenized_courses[:1]
```

```python
tokens_dict = gensim.corpora.Dictionary(tokenized_courses)
tokens_dict.token2id
```

```python
courses_bow = [tokens_dict.doc2bow(course) for course in tokenized_courses]
courses_bow[:1]
```

```python
bow_df.to_csv("../Data/bag of word.csv",index=False)
```
