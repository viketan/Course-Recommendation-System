### Plot a Word Cloud from Course Titles

```python
titles = " ".join(course_df['Title'])
```

```python
stopwords = stopwords.words('english')
```

```python
#stopwords.append(['using','course','online','class'])
```

```python
wordcloud = WordCloud(stopwords=stopwords, background_color="white", width=800, height=400)
```

```python
wordcloud.generate(titles)
```

```python
plt.figure(figsize=(10,10))
plt.tight_layout(pad=0)
plt.imshow(wordcloud, interpolation='bilinear')
plt.show()
```

### No. of courses by Instructor

```python
plt.figure(figsize=(20,10))
plt.xticks(rotation=90) 
counts = course_df['Instructor'].value_counts().sort_values(ascending=False)
ax = sns.countplot(x=course_df['Instructor'],order=counts.index)
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 5), textcoords = 'offset points')
```
