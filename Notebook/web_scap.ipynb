```python
import requests
from bs4 import BeautifulSoup as bs
import json
url = "https://ineuron.ai/courses"
```

```python
response = requests.get(url,verify=False)
```

```python
soup = bs(response.text,'html.parser')
```

```python
soup.prettify()
```

```python
# Find the script tag with the specified ID
script_tag = soup.find('script', id='__NEXT_DATA__')

# Get the content within the script tag
json_data = script_tag.string

# Load the JSON data
data = json.loads(json_data)
```

```python
title = list(data['props']['pageProps']['initialState']['init']['courses'].keys())[0]
```

```python
course_data = data['props']['pageProps']['initialState']['init']['courses']
title_list = []
desc_list = []
keyword_list = []
learn_list = []
instrutors_list = []
for course_name, course_details in course_data.items():
    description = course_details['description']
    if  len(course_details['instructorsDetails'])==0:
        instructor = None
    else:
        instructor = course_details['instructorsDetails'][0]['name']
    keyword = course_details['seo']['keywords'].replace("\n", ",")
    learn = course_details['courseMeta'][0]['overview']['learn']
    learn_string = ', '.join(learn)
    
    title_list.append(course_name)
    instrutors_list.append(instructor)
    keyword_list.append(keyword)
    learn_list.append(learn_string)
    desc_list.append(description)
```

```python
import pandas as pd
course_data = {'Title':title_list,'Instructor':instrutors_list,'Keywords':keyword_list,'Learn':learn_list,'Description':desc_list}
course_data = pd.DataFrame(course_data)
```
